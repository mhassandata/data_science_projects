```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load your data from a CSV file (replace 'your_data.csv' with the actual file name)
import csv

# Open the CSV file
import pandas as pd

# Load your data from an Excel file (replace 'Demodata.xlsx' with the actual file name)
data = pd.read_excel('Demodata.xlsx')

# Display the data
print(data)



# Select the features you want to use for clustering (assuming you have columns 'Feature1' and 'Feature2')
X = data[['Average', 'Exponential','Coefficient']]

# Number of clusters (K)
k = 3  # Adjust this based on your requirements

# Initialize and fit the KMeans model
kmeans = KMeans(n_clusters=k, random_state=0)
data['Class'] = kmeans.fit_predict(X)

# Visualize the clusters
for cluster_num in range(k):
    cluster_data = data[data['Class'] == cluster_num]
    plt.scatter(cluster_data['Average'], cluster_data['Exponential'], label=f'Class {cluster_num + 1}')

# Plot cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='black', marker='x', label='Cluster Centers')

plt.xlabel('Average')
plt.ylabel('Exponential')
plt.legend()
plt.title('Class Data')
plt.show()
```
